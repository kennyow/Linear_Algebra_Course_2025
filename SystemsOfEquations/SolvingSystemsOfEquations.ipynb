# System of Equations

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import display, clear_output  # Ensure correct import
import sympy as sp
```

## 2D

```python
eq1o = [1., 2, 1]  # [a, b, c]
eq2o = [2., 1, 3]

for i in range(10):
    plt.cla()  # Clear plot

    # Randomly update equations
    eq1 = np.add(eq2o, np.random.randn(1) * eq1o)
    eq2 = np.add(eq1o, np.random.randn(1) * eq2o)

    # Plot new lines (solutions)
    y1 = (np.array([eq1[1] * -3, eq1[1] * 3]) + eq1[2]) / eq1[0]
    plt.plot([-3, 3], y1, label="eq1")

    y2 = (np.array([eq2[1] * -3, eq2[1] * 3]) + eq2[2]) / eq2[0]  # Fixed denominator
    plt.plot([-3, 3], y2, label="eq2")

    plt.axis([-3, 3, -3, 6])
    plt.legend()

    # Pause to allow inspection
    clear_output(wait=True)  # Use the correct function
    display(plt.gcf())  # Display the updated plot
    time.sleep(0.1)
```

## 3D

```python
#these are the coefficientes of the equation:
# az = bx + cy + d
eq1o = [1, 2, 3, -1]  # [a, b, c, d]
eq2o = [2, 1, 3, 3]

# set up for 2D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for i in range(0, 10):
    #plt.cla()  # Clear plot

    # Randomly update equations
    eq1 = np.add(eq2o, np.random.randn(1) * eq1o)
    eq2 = np.add(eq1o, np.random.randn(1) * eq2o)

    # Plot new lines (solutions)
    y = (np.array([eq1[1] * -3, eq1[1] * 3]) + eq1[3]) / eq1[0]
    z = (np.array([eq1[2] * -3, eq1[2] * 3]) + eq1[3]) / eq1[0]  # Fixed denominator
    ax.plot([-3,3], y, z)

    # Plot new lines (solutions)
    y = (np.array([eq2[1] * -3, eq2[1] * 3]) + eq2[3]) / eq2[0]
    z = (np.array([eq2[2] * -3, eq2[2] * 3]) + eq2[3]) / eq2[0]  # Fixed denominator
    ax.plot([-3,3], y, z)

    #axus limits
    ax.set_xlim3d(-3, 6)
    ax.set_ylim3d(-3, 6)
    ax.set_zlim3d(-1, 10)

    

    # Pause to allow inspection
    clear_output(wait=True)  # Use the correct function
    display(plt.gcf())  # Display the updated plot
    time.sleep(0.1)
```

## Reduced Row Echelon Form

```python
#using sympy
# Criar algumas matrizes aleatrias
A = sp.Matrix(np.random.randn(4, 4))
B = sp.Matrix(np.random.randn(4, 3))

# Calcular a forma escalonada reduzida (RREF)
rrefA = A.rref()
rrefB = B.rref()

# Imprimir as matrizes RREF
print(np.array(rrefA[0]).astype(np.float64))  # Convertendo para float64 para melhor visualizao
print()
print(np.array(rrefB[0]).astype(np.float64))
```
